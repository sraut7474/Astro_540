```python
import numpy as np
import scipy.integrate as integrate
import matplotlib.pyplot as plt
```

```python
def E(x,Om,Ode,Or,Ok):
    return 1/(np.sqrt((Om*(1+x)**3)+(Ode*(1+x)**(3*(1-1)))+(Or*(1+x)**4)+(Ok*(1+x)**2)))

def R(y,Om,Ode, Or,Ok):
    return float((integrate.quad(E,0,y,args=(Om,Ode,Or,Ok)))[0])
```

```python
R(1.5,1,0,0,0)
```

```python
y=np.linspace(0,6,100)
z=[(1+a)*R(a,1,0,0,0) for a in y]
z1=[(1+a)*R(a,0,1,0,0) for a in y]
z2=[(1+a)*R(a,.3,.7,0,0) for a in y]
z3=[(1+a)*R(a,0,0,1,0) for a in y]
plt.plot(y,z,label="Einstein-de Sitter $\Omega_M=1$")
plt.plot(y,z1,label="$\Omega_\Lambda=1$")
plt.plot(y,z2,label="$\Omega_M=.3 \ \Omega_\Lambda=.7$")
plt.plot(y,z3,label="$\Omega_R=1$")
plt.ylim((0,18))
plt.legend()
plt.xlabel("Redshift (Z)")
plt.ylabel('Luminosity Distance H_0 d_L(z)')
plt.title("Luminosity Distance vs Redshift")
plt.figure(figsize=(1, 15),dpi=100)
plt.show()
```

```python
y=np.linspace(0,6,20)
z=[R(a,1,0,0,0)/(1+a) for a in y]
z1=[R(a,0,1,0,0)/(1+a) for a in y]
z2=[R(a,.3,.7,0,0)/(1+a) for a in y]
z3=[R(a,0,0,1,0)/(1+a) for a in y]
plt.plot(y,z,label="Einstein-de Sitter $\Omega_M=1$")
plt.plot(y,z1,label="$\Omega_\Lambda=1$")
plt.plot(y,z2,label="$\Omega_M=.3 \ \Omega_\Lambda=.7$")
plt.plot(y,z3,label="$\Omega_R=1$")
plt.xlabel("Redshift (Z)")
plt.ylabel('Angular Diameter Distance $H_0 d_A(z)$')
plt.title("Angular Diamter versus Redshift")
plt.legend()
plt.show()
```
